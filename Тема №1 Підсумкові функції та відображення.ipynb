## 3. Застосування відображень

```python
# Метод map() для заміни значень у колонці 
source_map = {'Nation-state': 'State-sponsored', 'Insider': 'Internal Threat', 'Hacker Group': 'Cybercriminal',
              'Unknown': 'Unknown Source'}
source_map = cyberattacks['Attack Source'].map(lambda s: source_map[s])
```

```python
source_map
```

```python
 # Метод map() для додавання нового стовпця 
cyberattacks['Loss Status'] = cyberattacks['Financial Loss (in Million $)'].map(
    lambda x: 'High Loss' if x > 40 else 'Low Loss')
cyberattacks
```

```python
def resolution_category(hours):
    if hours <= 24:
        return 'Fast'
    elif hours <= 50:
        return 'Medium'
    else:
        return 'Slow'
speed = cyberattacks['Incident Resolution Time (in Hours)'].apply(resolution_category)
speed
```

## 4 Вбудовані операції над стовпцями

### - Операція між двома стовпцями однакової довжини

## 5 Додаткові можливості бібліотеки pandas

```python
# Використання методу value_counts() у побудові графіків
cyberattacks['Year'].value_counts().sort_index().plot(kind='line')
```

```python
# Метод мода - найпоширінішого значення
most_common_attack = cyberattacks['Attack Type'].mode()[0]
most_common_source = cyberattacks['Attack Source'].mode()[0]
print(f"Найпоширеніший тип атаки: {most_common_attack}")
print(f"Найпоширеніше джерело атаки: {most_common_source}")
```
