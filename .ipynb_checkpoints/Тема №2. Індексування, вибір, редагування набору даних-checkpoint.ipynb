## 6 Умовний вибір

### 6.1 Проста умова 

```python
avg_loss = cyberattacks["Financial Loss (in Million $)"].mean()
print("Середні фінансові втрати =", avg_loss)
high_loss_attacks = cyberattacks.loc[cyberattacks["Financial Loss (in Million $)"] > avg_loss]
high_loss_attacks.head(10)
```

```python
# Додатковий приклад
ransomware_attacks = cyberattacks.loc[cyberattacks["Attack Type"] == "Ransomware"]
ransomware_attacks.iloc[:5,:3]
```

### 6.2 Складна умова

```python
sort_attacks = cyberattacks.loc[(cyberattacks["Country"] == "France") & (cyberattacks["Attack Type"] == 'SQL Injection'),
["Country", "Year", "Attack Type"]]
sort_attacks.head(10)
```

### 6.3 Умовні селектори

```python
# Приклад 1 умовний селектор is in
cyberattacks.loc[cyberattacks['Defense Mechanism Used'].isin(['VPN'])]
```

```python
cyberattacks['Defense Mechanism Used'].isin(['VPN'])
```

```python
# Приклад 2
cyberattacks.loc[cyberattacks['Year'].isin(['2015', '2016'])]
```

```python
# Приклад методу isnull()
missing_defense = cyberattacks.loc[cyberattacks['Defense Mechanism Used'].isnull()]
missing_defense.head()
```

```python
#  Метод notnull
sort_def = cyberattacks.loc[cyberattacks['Defense Mechanism Used'].notnull()]
sort_def.head()
```
