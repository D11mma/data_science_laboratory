```python
# Імпортування бібліотек
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
user_behavior = pd.read_csv("Datasets/user_behavior.csv")
```

```python
X = user_behavior[['Instagram visit score','Spending_rank(0 to 100)']].values
```

## Побудова моделі

```python
clalist = []
```

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, random_state=42)
    kmeans.fit(X)
    clalist.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), clalist, marker='o')
plt.title('Метод "Ліктя" для визначення')
plt.xlabel('Кількість кластерів')
plt.ylabel('WCSS (Sum of Squared Distances)')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters=4, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X)
print("Координати центроїдів кластерів:")
print(kmeansmodel.cluster_centers_)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 50, c = 'gray', label = 'Кластер 0: Неактивні')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 50, c = 'green', label = 'Кластер 1: Споживачі Контенту')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 50, c = 'magenta', label = 'Кластер 2: Premium')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 50, c = '#8E44AD', label = 'Кластер 3: Цінні Епізодичні')
plt.scatter(kmeansmodel.cluster_centers_[:, 0], kmeansmodel.cluster_centers_[:, 1], s = 200, c = 'yellow', label = 'Центроїди')
plt.title('Сегментація Користувачів', fontsize=16)
plt.xlabel('Оцінка відвідуваності', fontsize=14)
plt.ylabel('Рейтинг витрат', fontsize=14)
plt.legend()
plt.show()
```

## Опис кластерів: 
*Кластер 0* - даний сегмент складається з користувачів, які демонструють низьку оцінку відвідуваності та низький рейтинг витрат. Вони є пасивною аудиторією, яка рідко взаємодіє з платформою і має мінімальну фінансову цінність. Їхня низька залученість робить їх найменш пріоритетним об'єктом для дорогих маркетингових зусиль.

*Кластер 1* - цей сегмент характеризується високою оцінкою відвідуваності, що свідчить про їхню активну присутність, але при цьому низьким рейтингом витрат. Це "вільні" користувачі, які охоче споживають контент, але не готові платити. Вони важливі для охоплення та рекламного доходу.

*Класте 2* - найцінніший сегмент, який поєднує високу оцінку відвідуваності та високий рейтинг витрат. Вони є лояльними, активними та платоспроможними клієнтами. Основна стратегія щодо них — збереження лояльності та пропозиція ексклюзивних послуг, спрямованих на максимізацію їхньої життєвої цінності.

*Кластер 3* - сегмент має низьку оцінку відвідуваності, але високий рейтинг витрат. Це платоспроможні користувачі, які заходять рідко, але готові витрачати, коли це роблять. Їхня низька активність робить їх чудовим об'єктом для стратегій реактивації та персоналізованого ретаргетингу.


```python
silhouette_avg = silhouette_score(X, y_kmeans)
print(f"Коефіцієнт Силуету для K=4 становить: {silhouette_avg:.4f}")
```

```python
unique_labels, counts = np.unique(y_kmeans, return_counts=True)
cluster_sizes = dict(zip(unique_labels, counts))
print("Кількість користувачів у кожному кластері:")
print(f"Кластер 0 (Неактивні):               {cluster_sizes.get(0, 0)} користувачів")
print(f"Кластер 1 (Споживачі Контенту):      {cluster_sizes.get(1, 0)} користувачів")
print(f"Кластер 2 (Premium):                 {cluster_sizes.get(2, 0)} користувачів")
print(f"Кластер 3 (Цінні Епізодичні):        {cluster_sizes.get(3, 0)} користувачів")
```
